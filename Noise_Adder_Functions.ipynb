```
#example: Ender --> £nd£r
def change_e_to_euro_sign(str):


    for i in str:
        if i == 'E':
            str = str.replace('E', '€')
        if i == 'e':
            str = str.replace('e', '€')
    print(str)
change_e_to_euro_sign("ender")
```

```
#example: Kravat--> Gravat
def change_k_to_g(str):


    for i in str:
        if i == 'K':
            str = str.replace('K', 'G')
        if i == 'k':
            str = str.replace('k', 'g')
    print(str)

change_k_to_g("kravat")
```

```
# example: Risk --> riks/ekşi  --> eşki
def swap_consonants(str):
    str.casefold()
    j = 0
    for i in str:
        j = j + 1
        if j+1 > len(str)-1:
          break
        if i in 'aeiou' and str[j] in 'bcçdjklmnprsştvwxyz' and str[j + 1] in 'bcçdjklmnprsştvwxyz':
            temp = str
            temp2 = list(temp)
            temp2[j+1], temp2[j] = temp2[j], temp2[j+1]
            temp = ''.join(temp2)
            print(temp)
swap_consonants("kirpik")
```

```
# example: Göz göze --> Gözgöze
def remove_space_from_reiterative(str):
    str.casefold()
    for i in str:
        if i == ' ':
            str = str.replace(' ','')
            print(str)
remove_space_from_reiterative("yüz yüze")

#example: Ankara'da --> Ankarada
def remove_apostrophe_from_reiterative(str):

    for i in str:
        if i == "'":
           str = str.replace("'","")
           print(str)
remove_apostrophe_from_reiterative("Ankara'da")
```

```
# example: Rica --> Rici // Naber-->nobor // merhaba-->marhaba
def vowel_change_to_random_vowel(str):
    vowels = ['a', 'e', 'ı', 'i', 'o',
             'ö', 'u', 'ü']
    str.casefold()
    for i in str:
        if i in 'aeıioöuü':
            str = str.replace(i,vowels[random.randrange(len(vowels))])
            print(str)
            break

vowel_change_to_random_vowel("merhaba")
```

```
# example: teknosa-->tiquaa/fantastik-->fanwasrib
def randomly_change_word(str):

    inds = [i for i, _ in enumerate(str) if not str.isspace()]

    sam = random.sample(inds, 3)

    letts = iter(random.sample(ascii_letters, 3))
    lst = list(str)
    for ind in sam:
        lst[ind] = next(letts)

    print("".join(lst))
randomly_change_word("hello")
```

```
def disrupt_letter(word):#return s and ş to $
    word = word.casefold()
    if 's' in word:
        word = word.replace("s", '$')

    if 'ş' in word:   
        word = word.replace("ş", '$')

    return word 
```

```
def make_upper_case_vowels(word):#make some letters uppercase 
    
    x = random.randint(0,len(word)-1)
    
    for i in range(len(word)):
        word = word[:x] + word[x].upper() + word[x + 1:]
        
    return word
```

```
def increase_num_of_vowels(word):#increase the number of letter
    word = word.casefold()
    x = random.randint(2,5)
    y = random.randint(0,len(word)-1)
    
    for i in range(x):
        word = word[:y] + word[y] + word[y:]
    
    return word
```

```
def remove_vowels(word):#remove vowels in string
    word = word.casefold()
    vowels = ["a","e","ı","i","o","ö","u","ü"]
    
    for i in word:
        if i in vowels:
            word = word.replace(i, "")
    
    return word
```

```
def abbreviation(word):#bir tane --> bi tane
    word = word.casefold()
    for i in range(len(word)):
        if "bir" in word:
            start_index = word.find("bir")
            letter_index = start_index + 2
            word = remove_at(letter_index, word)
      
    return word
```

```
def turkish_letters(word):#return turkish characters to english characters
    word = word.casefold()

    if "ç" in word:   
        word = word.replace("ç", "c")
    
    if "ğ" in word:   
        word = word.replace("ğ", "g")
    
    if "ı" in word:   
        word = word.replace("ı", "i")
    
    if "ö" in word:   
        word = word.replace("ö", "o")
        
    if "ş" in word:   
        word = word.replace("ş", "s")
        
    if "ü" in word:   
        word = word.replace("ü", "u")

    return word 
```

```
def conjunctions(word):#write conjunctions contiguous
                       #ex: bana da mı --> banadamı
    word = word.casefold()
    
    if " de" in word:
        index = word.find(" de")
        word = remove_at(index, word)
    if " da" in word:
        index = word.find(" da")
        word = remove_at(index, word)
    if " ki" in word:
        index = word.find(" ki")
        word = remove_at(index, word)
    if " mi" in word:
        index = word.find(" mi")
        word = remove_at(index, word)
    if " mı" in word:
        index = word.find(" mı")
        word = remove_at(index, word)
        
    return word
```

```
def remove_at(i, word):#remove character at specific index
    return word[:i] + word[i+1:] 
```

```
def case_sensitivity(word):#write letters in lowercase instead of uppercase
                          #Ahmet --> ahmet
    for i, c in enumerate(word):
        if c.isupper():
            word = word.replace(c, c.lower())
    
    return word
```
